```python
#%matplotlib notebook
import scipy.misc
import scipy.spatial
import numpy as np
import matplotlib.pyplot as plt
```

```python
img  = scipy.misc.imread('toy.jpg')
trimap = scipy.misc.imread('toyTrimap.png', flatten='True')
h,w,c = img.shape
plt.imshow(img)
plt.show()
plt.imshow(trimap, cmap = 'gray')
plt.show()
```

```python
bg = trimap == 0
fg = trimap == 255
unknown = True ^ np.logical_or(fg,bg)
```

```python
fg_px = img[fg]
bg_px = img[bg]
```

```python
unknown_px = img[unknown]
```

```python
fg_hull = scipy.spatial.ConvexHull(fg_px)
fg_vertices_px = fg_px[fg_hull.vertices]

bg_hull = scipy.spatial.ConvexHull(bg_px)
bg_vertices_px = bg_px[bg_hull.vertices]
```

```python
d_fg = np.ones(unknown_px.shape[0]*fg_hull.equations.shape[0]).reshape(unknown_px.shape[0],fg_hull.equations.shape[0])*1000
for j, fg_eq in enumerate(fg_hull.equations):
    for i, unknown_px_val in enumerate(unknown_px):
        nhat= fg_eq[:3]
        d_fg[i,j] = np.dot(nhat,unknown_px_val) + fg_eq[3]
```

```python
d_bg = np.ones(unknown_px.shape[0]*bg_hull.equations.shape[0]).reshape(unknown_px.shape[0],bg_hull.equations.shape[0])*1000
for j, bg_eq in enumerate(bg_hull.equations):
    for i, unknown_px_val in enumerate(unknown_px):
        nhat= bg_eq[:3]
        d_bg[i,j] = np.dot(nhat,unknown_px_val) + bg_eq[3]
```

```python
d_bg1 = np.maximum(np.amax(d_bg,axis=1),0)
d_fg1 = np.maximum(np.amax(d_fg,axis=1),0)
```

```python
alpha = d_bg1/(d_bg1+d_fg1)
```

```python
unknown2 = unknown.astype(float).copy()
unknown2[unknown !=0] = alpha
plt.imshow(unknown2+fg, cmap='gray')
plt.show()
```

```python
plt.imshow(((unknown2+fg).reshape(h,w,1).repeat(3,2)*img).astype(np.uint8))
plt.show()
```
